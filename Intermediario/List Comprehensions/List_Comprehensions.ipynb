## O que  e qual a funo de list comprehension? 

 **List Comprehension**  uma estrutura de dados em Python.  bastante til no mundo pythonico por: 

-  permitir a construo de listas utilizando uma s **linha de cdigo**, de uma forma concisa e eficiente ;
- e por ser uma **alternativa** ao uso de loops  para criar listas.

Vamos supor que temos uma lista com diferentes tamanhos de DNA , mas queremos gerar uma **NOVA LISTA**  subtraindo os tamanhos de primer de cada sequncia da lista original.

**Primers**  so curtas sequncias de nucleotdeos inseridas junto da sequncia amostral. Para saber mais sobre primers: [link](https://knoow.net/ciencterravida/biologia/primer-ou-iniciador-molecular/)

```python
tamanhos_de_DNA = [18000, 22000, 26000, 28000, 31000, 33000, 35000, 38000, 42000]
primer = 30
tamanhos_DNA_sem_primer = []

for tamanho in tamanhos_de_DNA:
  novo_tamanho = tamanho - primer
  tamanhos_DNA_sem_primer.append(novo_tamanho)

print(tamanhos_DNA_sem_primer)
```

**No exemplo acima:**
- utilizamos **loop**  para percorrer a lista original;
- utilizamos o mtodo append() para adicionar o resultado da subtrao  **NOVA LISTA**  .

Mas e se existisse uma forma de reduzir 3 linhas deste cdigo? 

E existe!  **Vamos utilizar o list comprehensions nesse exemplo!**

## Sintaxe do list comprehension

- A estrutura  escrita dentro de colchetes j que estamos criando uma **NOVA LISTA** ;
- Toda a estrutura  escrita em apenas UMA linha de comando.

Calma  , vamos entender isso melhor por partes, **veja o exemplo citado utilizando list comprehensions**.  

```python
tamanhos_de_DNA = [18000, 22000, 26000, 28000, 31000, 33000, 35000, 38000, 42000]
primer = 30

tamanhos_DNA_sem_primer = [tamanho - primer for tamanho in tamanhos_de_DNA]
print(tamanhos_DNA_sem_primer)
```

Neste cdigo, o que de fato  list comprehension  a parte `[tamanho-primer for tamanho in tamanhos_de_DNA]`.


---



### **Preste ateno em como um list comprehensions  montado:**

1) Utilizamos a sintaxe bsica de um loop  : `for tamanho in tamanhos_de_DNA`;

2) Ao invs de colocar dois pontos e ir para a prxima linha, fazemos o que queremos com o loop **NA FRENTE**  do for: `tamanho - primer`;

3) O **resultado** dessa subtrao j ficar dentro da **NOVA LISTA**, identificada pelos colchetes [ ]. Voc percebeu que o list comprehensions foi feito dentro de colchetes? Ou seja, **j estamos dentro na NOVA LISTA** .

4) Ao final, atribumos essa NOVA LISTA que foi criada a uma varivel `tamanhos_DNA_sem_primer`

## Condicionais em list comprehensions

Vamos retornar ao nosso exemplo do tamanho de DNA:

```python
tamanhos_de_DNA = [18000, 22000, 26000, 28000, 31000, 33000, 35000, 38000, 42000]
primer = 30

tamanhos_DNA_sem_primer = [tamanho - primer for tamanho in tamanhos_de_DNA]
print(tamanhos_DNA_sem_primer)
```

Vamos supor que eu queira diminuir o primer APENAS se o tamanho do DNA foi maior que 30.000. 

**Como fazemos isso com list comprehensions?** 

```python
tamanhos_de_DNA = [18000, 22000, 26000, 28000, 31000, 33000, 35000, 38000, 42000]
primer = 30

tamanhos_DNA_sem_primer = [tamanho - primer for tamanho in tamanhos_de_DNA if tamanho>30000]
print(tamanhos_DNA_sem_primer)
```

 **Compare ambos os cdigos, o que eles possuem de diferente?**

A nica parte de diferente na escrita do cdigo foi o `if tamanho>30000`, o qual tambm  escrito na mesma linha.

**Para adicionar uma condicional ao list comprehensions, basta coloc-la ao final, aps o loop.**

#### Else

Caso voc utilize um **else**,  necessrio colocar a frente do for 

```python
tamanhos_de_DNA = [18000, 22000, 26000, 28000, 31000, 33000, 35000, 38000, 42000]
primer = 30

tamanhos_DNA_sem_primer = [tamanho - primer if tamanho>30000 else None for tamanho in tamanhos_de_DNA ]
print(tamanhos_DNA_sem_primer)
```
