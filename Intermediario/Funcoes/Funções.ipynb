**Exemplo 1** 

Uma funo chamada `verificar_proteina` que recebe como parmetro uma sequncia de aminocidos, este parmetro foi denominado `sequencia_aminos`. 

A funo verifica se essa sequncia tem:

1) **comprimento menor que 10** ;

2) **o primeiro caracter da sequncia  diferente de um **.

Do ponto de vista biolgico, o primeiro caracter  um M na maioria das protenas dos seres vivos , pois simboliza o aminocido **Metionina**. 

Para saber mais sobre a Metionina e a traduo de protenas : [link](https://pt.khanacademy.org/science/biology/gene-expression-central-dogma/translation-polypeptides/a/the-stages-of-translation)

````python
def verificar_proteina(sequencia_aminos):
  if len(sequencia_aminos)<10 or sequencia_aminos[0]!="M":
    print("Est protena no  vlida")
  else:
    print("A protena  vlida")
````

**Exemplo 2** 

Uma funo que se chama `is_nucleotide_valid` e recebe como parmetro `nucleotide`.

A funo mostra se o caracter associado ao parmetro nucleotide  um nucleotdeo vlido, ou seja,  A ou C ou T ou G

````python
def is_nucleotide_valid(nucleotide):
  nucleotides_list = ['A', 'C', 'T', 'G']
  nucleotide = nucleotide.upper()
  
  if nucleotide in nucleotides_list:
    print(f"{nucleotide}  um nucleotdeo vlido")
  else:
    print(f"{nucleotide} NO  um nucleotdeo vlido")
````

## Return

Poderamos utilizar a **instruo return** ao invs de sempre mostrarmos na tela.

A instruo **return** finaliza a execuo e retorna um valor da funo.

 recomendado utilizar **return** ao invs de **print()** em cdigos maiores.

Vamos substituir o **Exemplo 2** para retornar `True` ou `False`

````python
def is_nucleotide_valid(nucleotide):
  nucleotide = nucleotide.upper()
  nucleotide_list = ['A', 'C', 'T', 'G']
  if nucleotide in nucleotide_list:
    return True
  else:
    return False
````

No dia a dia, na construo de scripts e pipeline, ** muito comum utilizar return com os valores booleanos**.

**Vamos pensar em um exemplo:** digamos que queremos verificar a quantidade de um nucleotdeo especfico em um arquivo FASTA* e para isso precisaramos verificar que o nucleotdeo que vamos procurar  vlido. 

Para este script que precisamos construir, faz mais sentido que a funo retorne `True` ou `False` e assim o script siga ou pare de acordo com a "resposta".

*arquivo FASTA :  um tipo de arquivo muito comum no mundo da Bioinformtica. Para saber mais sobre este: [link](https://blog.varsomics.com/fasta-como-realizar-a-leitura-de-sequencias/#:~:text=O%20arquivo%20FASTA%20codifica%20um,listas%20de%20sequ%C3%AAncias%20de%20prote%C3%ADnas.)


````python
caracter = input("Digite um nucleotdeo para verificao: ")
if is_nucleotide_valid(caracter):
  print("Abrindo o arquivo...")
  #abre o arquivo
else:
  print("O arquivo no ser aberto, pois o caracter no  um nucleotdeo")
````

##  possvel definir parmetros padres

Assim, se nenhum parmetro for passado, mas a funo for executada, **o Python j tem um valor pr-definido**.

Para fazer isso, ao criar a funo basta colocar o parmetro com um "=" e um valor ao lado. Observe a funo que j criamos obtendo "**A**" como o valor padro de `nucleotide`

````python
def is_nucleotide_valid(nucleotide='A'):
  nucleotide = nucleotide.upper()
  nucleotide_list = ['A', 'C', 'T', 'G']
  if nucleotide in nucleotide_list:
    return True
  else:
    return False
````

Assim, ao executarmos a funo sem passar um parmetro, "A"  executado

````python
is_nucleotide_valid()
````
