# Expresses aritmticas

Uma expresso aritmtica  formada por nmeros (chamados de operandos) e operadores.
Os operadores podem realizar soma (+), subtrao (-), diviso (/) ou multiplicao (*) dos operandos.

## Precedncia e associatividade

As regras de precedncia indicam qual operador  calculado primeiro. Por exemplo, qual o resultado da expresso 2 - 3 * 4?

Como a multiplicao tem maior prioridade que a subtrao, o produto 3*4  reduzido ao valor 12 e a seguir se calcula o valor da subtrao 2 - 12, resultando em -10.

As regras de associatividade indicam a ordem dos clculos para operadores que tenham a mesma precedncia.
