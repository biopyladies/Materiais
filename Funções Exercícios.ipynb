⭐ **DESAFIO:** Implemente uma função em Python que calcule o **modelo exponencial** de declínio da população. 

O modelo leva em conta apenas a taxa de mortalidade, representada pelo parâmetro `rate`. A função deve receber como parâmetros: `initial_population`, `rate`, `years`.

A função deve retornar a população final.


---


**EXEMPLO:**

initial_population = 100

rate = 0.1

years = 5


A população final deve ser 60.6

**DICA:** 🤓

O modelo exponencial da população segue uma fórmula: P(t) = P0 * e^(-rt) 

* P(t) é a população no tempo t (em anos).
* P0 é a população inicial.
* e é a base dos logaritmos naturais, aproximadamente igual a 2.718.
* r é a taxa de crescimento ou declínio da população (por ano).
* t é o número de anos decorridos desde a população inicial.

no exemplo dado: P(t) = 100 * e^(-0.1*5)
